### Scope of variable 
##### Types : Global scope and Local scope
##### variables which are accissible outside and inside the function are known as global variable

```python
a = lambda x:x+10
a(2)
```

```python
x = 10
def abc():
    print(x)

abc()
print(x)
```

```python
#how to change global variable value inside the function
x = 10 
def abc():
    global x
    print(x)
    x = 30
    print(x)
```

```python
abc()
print(x)
```

```python
y = 30 
def abc():
    global y
    y = 20
    print(y)
```

```python
abc()
print(y)
```

### map function
#### map function always takes two arguments 1st is function and second one is data type
#### if we want to see the output we need to typecast it into list
##### (use map function when we want to perform same type of operation on complete set)

```python
l = [4,53,24,2,9]
z = list(map(lambda x:x**2 , l))
print(z)
```

```python
l = [10,20,30,40]
z = list(map(lambda x:x+10,l))
z
```

```python
l = ["maseera","sahil","cds","sssss"]
z = list(map(lambda x : len(x),l))
z
```

```python
l = ["maseera","sahil","cds","sssss"]
z = set(map(lambda x : len(x),l))
z
#duplicate lengths are neglected in set
```

### Filter
##### used while filtering out elements

```python
l = [1,2,387,6,10,15,20]
even = list(filter(lambda x : x%2==0,l))
even
```

```python
l = [1,2,387,6,10,15,20]
odd = list(filter(lambda x : x%2 != 0,l))
odd
```

```python
l = ["abc","xyz",1,2,3.4,6,5]
s = list(filter(lambda x: type(x)==str,l))
i = list(filter(lambda x: type(x)==int,l))
f = list(filter(lambda x: type(x)==float,l))
print(s)
print(i)
print(f)
```

```python
def abc(n):
    if n == 1:
        return 1
    else:
        return n + abc(n-1)

abc(5)
```

```python
def xyz(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * xyz(n-1)

xyz(5)
```

```python
def fe(n):
    if n == 0 or n == 1:
        return 1
    else:
         return fe(n-2) + fe(n-1)
```

```python
fe(4)
```

```python
def fs(n):
        if n == 0 or n == 1:
            return 1
        else:
            return fs(n-2) + fs(n-1)

n = int(input("enter length"))
if n == 0:
         print(1)
else:
    for i in range(n):
        print(fs(i),end=" ")
```

```python
fs(6)
```
